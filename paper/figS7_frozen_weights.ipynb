```python
import matplotlib.pyplot as plt
import tools.dataTools as dt
from constants import Constants
import figure_params
figure_params.set_rc_params()
import os
figure_params.setup_directories()

SUBFIG_DIR = Constants.FIG_FOLDER + 'figS7_frozen_weights/'
if not os.path.exists(SUBFIG_DIR):
    os.makedirs(SUBFIG_DIR)

#run simulations for figure
import subprocess
subprocess.call(['sh', '../simulation/fig_simulations/Sfig_fixed_layers_sims.sh'])
```

```python
repertoires = Constants.UNIS
seed_ex = Constants.SEED_EX
sim_set = 'uni_rad'
```

#### Frozen input weights

```python
# rotation
# _, params = dt.graph_outputs_all(seed_ex, sim_set, repertoires, 
#     perturbation = 'rotation', pert_param = 10.0 ,pert_set = 'v3')
_,_, params = dt.graph_loss_broken_axis(Constants.SEEDS, sim_set, repertoires, 'rotation',10.0 , 'v3', smooth = True)
plt.savefig(SUBFIG_DIR + "loss_"+params+".pdf", format='pdf', bbox_inches='tight' )

# _, params = dt.graph_outputs_all(seed_ex, 'uni_onehot', repertoires, 
#     perturbation = 'rotation', pert_param = 10.0 ,pert_set = 'v3')
_,_, params =dt.graph_loss_broken_axis(Constants.SEEDS, 'uni_onehot', repertoires, 'rotation',10.0 , 'v3', smooth = True)
plt.savefig(SUBFIG_DIR + "loss_"+params+".pdf", format='pdf', bbox_inches='tight' )
```

```python
#reassoc, frozen inputs
# _, params = dt.graph_outputs_all(seed_ex, 'uni_onehot', repertoires[1:], 
#     perturbation = 'reassoc', pert_param = 0.0 ,pert_set = 'v3')

fig, ax = plt.subplots(ncols = 1, figsize = (3,3))
_, params =dt.graph_loss(Constants.SEEDS, 'uni_onehot', repertoires[1:], 'reassoc', 0.0, 'v3', 
    smooth=True, rel_loss = False, ax = ax, xlim=[0,300], colors = Constants.REPERTOIRE_COLORS[1:])
ax.hlines([0.25],xmin = 0, xmax = 300, linestyles='dashed', color = 'k')
plt.savefig(SUBFIG_DIR + "loss_"+params+".pdf", format='pdf', bbox_inches='tight' )

fig, ax = plt.subplots(ncols = 1, figsize = (3,3))
_, params =dt.graph_loss(Constants.SEEDS, 'uni_rad', repertoires[1:], 'reassoc', 0.0, 'v3', 
    smooth=True, rel_loss = False, ax = ax, xlim=[0,300], colors = Constants.REPERTOIRE_COLORS[1:])
ax.hlines([0.25],xmin = 0, xmax = 300, linestyles='dashed', color = 'k')
plt.savefig(SUBFIG_DIR + "loss_"+params+".pdf", format='pdf', bbox_inches='tight' )
```
